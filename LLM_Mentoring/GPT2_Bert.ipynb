#텍스트 생성 (GPT-2)


```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer #토큰 아이징
import torch
```

## 모델과 토큰나이저 로드

```python
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

## 프롬프트 입력

```python
prompt = "The future of robotics is"
inputs = tokenizer(prompt, return_tensors="pt")
```

## 텍스트 생성

```python
with torch.no_grad():
  outputs = model.generate(
      inputs["input_ids"],
      max_length = 100,
      num_return_sequences = 1,
      temperature = 0.7, #높으면 높을수록 어휘력이 풍부해진다 , 법률 전문가 이런애들은 답이많으면 안되니까 temp를 낮게하고 이런식
      top_p = 0.9,
      do_sample = True
  )
```

```python
generate_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generate_text)
```

## 감정 분석

```python
from transformers import BertForSequenceClassification, BertTokenizer
import torch
```

## 사전 학습된 모델 로드

```python
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(
    model_name,
    num_labels = 2 #긍정 or 부정
)
```

```python
text = "everyone hurry up!! you're so late class. teacher so upset"
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

with torch.no_grad():
  outputs = model(**inputs)
  predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)


print(f"긍정: {predictions[0][1].item(): .4f}")
print(f"부정: {predictions[0][0].item(): .4f}")
```
